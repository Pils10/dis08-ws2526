# parse.ipynb

Regex-based parsing of orders.csv

```
import re

with open("./csv/orders.csv", "r", encoding="utf-8") as f:
    text = f.read()

text[:300]
```

```
order_number_regex = r'[A-F0-9]{8}(?:-[A-F0-9]{4}){4}'
order_numbers = re.findall(order_number_regex, text)
order_numbers
```

```
product_code_regex = r'\b[A-Z0-9]{11}\b'
product_codes = re.findall(product_code_regex, text)
product_codes
```

```
price_regex = r'\$\d+\.\d{2}'
prices_raw = re.findall(price_regex, text)
prices = [float(p.replace("$", "")) for p in prices_raw]
prices
```

```
date_regex = r'\b\d{2}/\d{2}/\d{4}\b'
dates = re.findall(date_regex, text)
dates
```

```
[p for p in prices if p > 500]
```

```
def swap_date(match):
    mm, dd, yyyy = match.group(1), match.group(2), match.group(3)
    return f"{dd}/{mm}/{yyyy}"

re.sub(r'(\d{2})/(\d{2})/(\d{4})', swap_date, text).splitlines()[1:4]
```

```
items_regex = r',(\d+),\d{2}/\d{2}/\d{4}$'
items = [int(x) for x in re.findall(items_regex, text, re.MULTILINE)]
max(items)
```

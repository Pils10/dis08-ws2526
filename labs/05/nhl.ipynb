```
import pandas as pd

df = pd.read_csv('data.csv')

# Most wins
results = {}
for yr in [1990,2000,2010]:
    sub = df[df.year==yr]
    if not sub.empty:
        max_team = sub.loc[sub.wins.idxmax(), 'name']
        max_wins = sub['wins'].max()
        results[yr] = (max_team, max_wins)
    else:
        results[yr] = None

# number teams
counts = {yr: df[df.year==yr].shape[0] for yr in [1991,2001,2011]}

print("Most wins:", results)
print("Team counts:", counts)
```
